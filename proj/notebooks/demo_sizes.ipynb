```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
plt.plot([7732, 5019, 3608, 285])
plt.plot([4941, 5007, 3449, 145])
plt.xticks([0,1,2,3], ['DRD2','LogP','JAK2-max','JAK2-min'])
plt.yticks([0, 2000, 4000, 6000, 8000])
plt.xlabel('Experiment')
plt.ylabel('Num. of demonstrations')
plt.legend(['Unique compounds', 'Compounds in threshold'])
sns.despine()
plt.savefig('demo_sizes.pdf')
```

```python
df = pd.DataFrame({'x':[1,2,3,4,5,6,7,8], 
                   'exp':['DRD2','LogP','JAK2-max','JAK2-min','DRD2','LogP','JAK2-max','JAK2-min'], 
                   'category':['Unique compounds','Unique compounds','Unique compounds','Unique compounds',
                              'Compounds in threshold','Compounds in threshold','Compounds in threshold',
                               'Compounds in threshold'],
                  'y':[7732, 5019, 3608, 285, 4941, 5007, 3449, 145]})
df.head(10)
```

```python
sns.catplot(x="exp", y="y", hue="category", kind="bar", data=df);
plt.xlabel('Experiment')
plt.ylabel('Num. of demonstrations')
sns.despine(offset=5)
plt.savefig('demo_sizes.eps')
```
